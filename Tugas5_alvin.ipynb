- SEQUENTIAL_SEARCH

```python
def sequential_search(A, x):
    print("Langkah-langkah Sequential Search:")
    for i in range(len(A)):
        print(f"Langkah {i+1}: Bandingkan x = {x} dengan A[{i}] = {A[i]}")
        if A[i] == x:
            print(f"--> Data ditemukan pada langkah {i+1}, indeks {i}\n")
            return i
    print("--> Data tidak ditemukan\n")
    return -1
A = [10, 50, 30, 70, 80, 20]
x = 30
sequential_search(A, x)
```

- INSEARTION_SORT

```python
def insertion_sort_pass3(A):
    print("Proses Insertion Sort hingga Pass ke-3:")

    for i in range(1, 4):  # Pass 1 sampai Pass 3
        key = A[i]
        j = i - 1

        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j -= 1

        A[j + 1] = key
        print(f"Setelah Pass {i}: {A}")

    return A
A = [64, 34, 25, 12, 22, 11, 90]
insertion_sort_pass3(A)
```
